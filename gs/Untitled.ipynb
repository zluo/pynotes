```python
df=dl('amd',100)  # download 100 days data 

# add 5, 10, 15 SMA
df['Sma5'] = df['Close'].rolling(5).mean() 
df['Sma10'] = df['Close'].rolling(10).mean()
df['Sma15'] = df['Close'].rolling(15).mean()

df
```
