```python
from os import listdir
import numpy as np
import operator

# Read file to the array

dict ={}

def readData(dir):
    fileLists = listdir(dir)
    for a in fileLists:
        key = a[:-4]  # extract symbol from file name (AMB.csv)
        mat, labels = file2matrix(a)
        dict[key] = mat


        
def file2matrix(filename):
    fr = open(filename)
    n = len(fr.readlines())
    mat = np.zeros([n, 13])
    labels = []
    fr = open(filename)
    i = 0
    for line in fr.readlines():
        line = line.strip()
        list = line.split(',')
        mat[i, :] = list[1:]
        #labels.append(int(list[-1]))
        labels.append(list[0]))
        i += 1
    return mat, labels        
```
