```python
import numpy as np
import operator
map ={'largeDoses':'1', 'smallDoses':'2','didntLike':3}

def file2matrix(filename):
    fr = open(filename)
    n = len(fr.readlines())
    mat = np.zeros([n, 3])
    labels = []
    fr = open(filename)
    i = 0
    for line in fr.readlines():
        line = line.strip()
        list = line.split('\t')
        mat[i, :] = list[0:3]
        #labels.append(int(list[-1]))
        labels.append(int(map.get(list[-1])))
        i += 1
    return mat, labels
                         
```

```python
# near k elements
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = np.tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortedDistIndicies = distances.argsort()     
    classCount={}          
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1
    sortedClassCount = sorted(classCount.iteritems(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
```

```python
#Normalize function

def autoNorm(dataSet):
    minVals = dataSet.min(0)  # min values of each column
    maxVals = dataSet.max(0)
    ranges = maxVals - minVals
    normDataSet = np.zeros(np.shape(dataSet))
    m = dataSet.shape[0]
    normDataSet = dataSet - np.tile(minVals , (m, 1))
    normDataSet = normDataSet / np.tile(ranges, (m,1))
    return normDataSet, ranges, minVals
    
    
```

```python
m,l = file2matrix('datingTestSet.txt')
print (m.min(0))
print (m.max(0))
print (m.shape)
n, r, minVals = autoNorm(m)
```

```python
# Test classifier
def datingClassTest():
    ratio = 0.10
    mat, labels = file2matrix('datingTestSet.txt')
    normMat, ranges, minVals = autoNorm(mat)
    m = normMat.shape[0]
    tn = int(m*ratio)
    errorCount = 0.0
    for i in range(tn):
        res = classify0(normMat[i], normMat[tn:m], labels[tn:], 3)
        if (res != labels[i]):
            print ("classifier Result: %d , Actual Result: %d" % (res, labels[i]))
            errorCount +=1.0
    print ("Error Rate: %f" % (errorCount/float(tn)))
               
```

```python
datingClassTest()
```
