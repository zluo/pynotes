### Matplotlib Basic


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(6)
```

### Basic plot input

```python
%matplotlib inline    
plt.plot(x,x**2)      # plot(x, y), it says, give me an array of x axis data, and give me an array of y axis data
plt.show()            # show
```

### How to set up graph size

```python
plt.figure(figsize=(10,8))
plt.plot(x,x**2)      # plot(x, y), it says, give me an array of x axis data, and give me an array of y axis data
plt.show()
```

### Multiline Plots

Display multiple lines in the same figures

```python
plt.plot(x,x**2)      
plt.plot(x,x*2) 
plt.plot(x,np.sqrt(x))  
plt.show()            # show
```

### Another format of multilines
plt.plot(x,y,x,y,...)


```python
plt.plot(x,x**2,x,x*2 ,x,np.sqrt(x))      
plt.show()       
```

### Add a grid

```python
plt.grid(True)
plt.plot(x,x**2,x,x*2 ,x,np.sqrt(x))      
plt.show()    
```

### Handle axes
plt.axis([xmin,xmax,ymin,ymax])

or

plt.xlim([xmin,xmax])
plt.ylim([ymin,ymax])

```python
plt.axis() # showing current axis limitations
plt.axis([0,10,0,10])
plt.show()
```

### Adding labels, title and legend

```python
%matplotlib inline    
import numpy as np
x = np.arange(1,5)
plt.plot(x,x*x,x,x*1.5,x,x/3.0)     
plt.grid(True)
plt.axis([0,5,0,10])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Title')
plt.legend(loc =2)
plt.legend(['Normal','Fast','Slow'])
plt.show()    
```
