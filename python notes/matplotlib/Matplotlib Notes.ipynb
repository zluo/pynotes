### Multiline Plots

I want to display multiple lines in the same figures

```python
%matplotlib inline    
import math
# how to import data
x = range(6)
plt.plot(x,[xi*xi for xi in x])      
plt.plot(x,[xi*2 for xi in x]) 
plt.plot(x,[math.sqrt(xi) for xi in x])  
plt.show()            # show
```

### Another format of multilines
plt.plot(x,y,x,y,...)


```python
%matplotlib inline    
import math
# how to import data
x = range(6)
plt.plot(x,[xi*xi for xi in x],x,[xi*2 for xi in x],x,[math.sqrt(xi) for xi in x])      
plt.show()       
```

### Add a grid

```python
%matplotlib inline    
import numpy as np
x = np.arange(1,5)
plt.plot(x,x*x,x,x*1.5,x,x/3.0)     
plt.grid(True)
plt.show()    
```

### Adding labels, title and legend

```python
%matplotlib inline    
import numpy as np
x = np.arange(1,5)
plt.plot(x,x*x,x,x*1.5,x,x/3.0)     
plt.grid(True)
plt.axis([0,5,0,10])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Title')
plt.legend(loc =2)
plt.legend(['Normal','Fast','Slow'])
plt.show()    
```
